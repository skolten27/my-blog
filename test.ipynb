```python
from plotnine import *
from palmerpenguins import load_penguins

penguins = load_penguins()
penguins_head = penguins.head()
html_table = penguins_head.to_html(index=False)

# Save the HTML table to a file or print it to copy-paste into your blog
with open("penguins_table.html", "w") as f:
    f.write(html_table)

plot = (ggplot(penguins,
               aes(x='bill_depth_mm',
                   y='bill_length_mm',
                   color='species'))
        + geom_point()
        + labs(
            title="Bill depth and length",
            subtitle="Dimensions for Adelie, Chinstrap, and Gentoo Penguins",
            x="Bill depth (mm)", y="Bill length (mm)",
            color="Species",
            caption="Source: Palmer Station LTER / palmerpenguins package"
        )
        + scale_color_manual(values=['#FC766AFF', '#B0B8B4FF', '#184A45FF'])  # Custom color scheme
        + theme_minimal())

# Display the plot (use print() in a script)
print(plot)
#plot.save('penguins_plot.png', width=10, height=7, dpi=400)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
from palmerpenguins import load_penguins


penguins = load_penguins()

plt.clf()
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=penguins,
    x='bill_depth_mm',
    y='bill_length_mm',
    hue='species',
    palette=sns.color_palette(['#FC766AFF', '#B0B8B4FF', '#184A45FF'])
)

plt.title('Bill Depth and Length', fontsize=16)
plt.suptitle('Dimensions for Adelie, Chinstrap, and Gentoo Penguins', fontsize=12)
plt.xlabel('Bill Depth (mm)', fontsize=14)
plt.ylabel('Bill Length (mm)', fontsize=14)

plt.figtext(0.9, 0.02, 'Source: Palmer Station LTER / palmerpenguins package', 
            horizontalalignment='right', fontsize=10)

# Show the plot
plot.save('penguins_plot_seaborn.jpg', width=10, height=7, dpi=400)
plt.show()
plt.clf()
```

```python
import matplotlib.pyplot as plt
from palmerpenguins import load_penguins

# Load the penguins dataset
penguins = load_penguins()

# Define the custom color map for each species
plt.clf()
color_map = {
    'Adelie': '#FC766AFF',  # Blue
    'Chinstrap': '#B0B8B4FF',  # Green
    'Gentoo': '#184A45FF'  # Red
}

# Map colors based on the species
colors = penguins['species'].map(color_map)
plt.figure(figsize=(10, 8))
# Create the scatter plot
plt.scatter(penguins['bill_depth_mm'], penguins['bill_length_mm'], c=colors)

# Add titles and axis labels
plt.title('Bill Depth and Length', fontsize=16)
plt.suptitle('Dimensions for Adelie, Chinstrap, and Gentoo Penguins', fontsize=12)
plt.xlabel('Bill Depth (mm)', fontsize=14)
plt.ylabel('Bill Length (mm)', fontsize=14)

# Create the legend manually
for species, color in color_map.items():
    plt.scatter([], [], c=color, label=species)

plt.legend(title='Species', loc='upper left')

# Add the caption
plt.figtext(0.9, 0.02, 'Source: Palmer Station LTER / palmerpenguins package', 
            horizontalalignment='right', fontsize=10)

# Adjust layout
plot.save('penguins_plot_matpltlib.jpg', width=10, height=7, dpi=400)
# Show the plot
plt.show()
```
